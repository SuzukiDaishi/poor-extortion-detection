```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
csv = pd.read_csv('西山貧乏ゆすりデータ.csv')
x, y, z, out = csv[['X', 'Y', 'Z', 'Out']].values.T
```

```python
def slide(data, f, width=5) :
    dist = []
    pad_data = np.pad(data, (0, width-1), 'mean')
    for i, _ in enumerate(data):
        dist.append(f(pad_data[i:i+width]))
    dist = np.array(dist)
    return dist
def binarization(data, threshold):
    dist = data.copy()
    dist[data < threshold]  = 0.
    dist[data >= threshold] = 1.
    return dist
def get_leg_vibration(x, y):
    """
    貧乏ゆすりを検出する(精度: 91%程度)
    
    param: 
        x: ndarray型で加速度センサXの取得
        y: ndarray型で加速度センサYの取得
    return: 
        ndarray型で1なら貧乏ゆすり、0ならそれ以外
    """
    mean_xy = np.mean([x, y], axis=0)
    slide_var_xy = slide(mean_xy,lambda x:x.var(),width=33)
    return binarization(slide_var_xy, 0.024)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.plot(x, out, 'o')
```

```python
clf = LinearRegression()

print('単回帰分析 x')
clf.fit(x.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_  
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(x.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 y')
clf.fit(y.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_  
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(y.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 z')
clf.fit(z.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_  
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(z.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 slide_var_xyz')
mean = np.mean([x, y, z], axis=0)
slide_var = slide(mean,lambda x:x.var(),width=33)
clf.fit(slide_var.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_ 
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(slide_var.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 slide_var_xy')
mean = np.mean([x, y], axis=0)
slide_var = slide(mean,lambda x:x.var(),width=33)
clf.fit(slide_var.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_ 
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(slide_var.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 slide_var_xz')
mean = np.mean([x, z], axis=0)
slide_var = slide(mean,lambda x:x.var(),width=33)
clf.fit(slide_var.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_ 
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(slide_var.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
print()

print('単回帰分析 slide_var_yz')
mean = np.mean([y, z], axis=0)
slide_var = slide(mean,lambda x:x.var(),width=33)
clf.fit(slide_var.reshape(-1, 1), out.reshape(-1, 1))
a = clf.coef_
b = clf.intercept_ 
print("回帰係数:", a)
print("切片:", b) 
print("決定係数:", clf.score(slide_var.reshape(-1, 1), out.reshape(-1, 1)))
print(f'y = {a[0,0]}*x + {b[0]}')
```

```python
mean = np.mean([x, y], axis=0)
slide_var = slide(mean,lambda x:x.var(),width=33)
clf.fit(slide_var.reshape(-1, 1), out.reshape(-1, 1))
plt.plot(clf.predict( slide_var.reshape(-1, 1) ))
plt.plot(slide_var )
```

```python
opt_threshold = []
for i in np.arange(0.0, clf.predict(slide_var.reshape(-1, 1)).max(), 0.001):
    opt_threshold.append( ( binarization(clf.predict(slide_var.reshape(-1, 1)).T[0], i)==out).sum() )
plt.title('Threshold accuracy')
plt.plot(opt_threshold)
```

```python
print('max threshold:', np.argmax(opt_threshold)*0.001, '\nacc:', np.max(opt_threshold) / slide_var.shape[0] )
```

```python
plt.plot(binarization(clf.predict(slide_var.reshape(-1, 1)).T[0], 0.333))
plt.plot(out)
```

```python
def noiseRemover(data, n):
    w = data.copy()
    for i in range(n):
        w = slide(w, lambda x: 0 if x[0]==1 and x[1]==0 or x[0]==0 and x[1]==1 else x[0] , width=2)
    for i in range(n):
        w = slide(w, lambda x: 1 if x[0]==1 and x[1]==0 or x[0]==0 and x[1]==1 else x[0] , width=2)
    return w
dist = binarization(clf.predict(slide_var.reshape(-1, 1)).T[0], 0.333)
n = []
for i in range(50):
    n.append( (noiseRemover(dist, i)==out).sum() ) 
plt.plot(n)
```

```python
n_max = np.argmax(n)
print('max noize filter n 1番目:', n_max, ', acc:', np.max(n) / out.shape[0] )
```

```python
plt.plot( out )
plt.plot( noiseRemover(dist, n_max) )
plt.show()
```
