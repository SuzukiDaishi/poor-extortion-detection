# 西山くんのやった方法でやってみる

## やり方

使用する変数を選択 → スライドで分散を求める(フィルタ幅を調査) → 二値化(閾値を調査)

## 調査方法

- 使用する変数
    - 分散を求め大きいものを調べる(そのほうが違いが出やすいと考えたため)
    - 全ての組み合わせの共分散を調べる(無駄なデータを省くため)
    - 実際に見て調査
- フィルタ幅の調査
    - フィルタ幅ごとの出力と教師データの共分散を最大化する組み合わせを調査
- 二値化の閾値を調査
    - 閾値ごとの出力と教師データのハミング距離を最大化する組み合わせを調査

## 結論

使用した変数: x, y <br>
フィルタ幅: 33 <br>
閾値: 0.024

91%程度を実現

## 補足

ノイズフィルタ(収縮膨張幅37)をかけると<br>
93%になった

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv = pd.read_csv('西山貧乏ゆすりデータ.csv')
```

```python
print(' - - var - -')
csv.var()
```

```python
print(' - - std - -')
csv.std()
```

```python
csv['X'].plot()
csv['Y'].plot()
csv['Z'].plot()
csv['Out'].plot()
plt.title('Visualization of CSV data')
plt.legend()
```

```python
x, y, z, out = csv[['X', 'Y', 'Z', 'Out']].values.T
```

```python
print('cov(x,y):', np.cov(x, y)[0,1])
print('cov(x,z):', np.cov(x, z)[0,1])
print('cov(y,z):', np.cov(y, z)[0,1])
```

```python
mean_xyz = np.mean([x, y, z], axis=0)
mean_xy = np.mean([x, y], axis=0)
mean_xz = np.mean([x, z], axis=0)
mean_yz = np.mean([y, z], axis=0)
plt.title('average of x, y and z')
plt.plot(mean_xyz)
plt.plot(out)
plt.show()
plt.title('average of x and y')
plt.plot(mean_xy)
plt.plot(out)
plt.show()
plt.title('average of x and z')
plt.plot(mean_xz)
plt.plot(out)
plt.show()
plt.title('average of y and z')
plt.plot(mean_yz)
plt.plot(out)
plt.show()
```

```python
def slide(data, f, width=5) :
    dist = []
    pad_data = np.pad(data, (0, width-1), 'mean')
    for i, _ in enumerate(data):
        dist.append(f(pad_data[i:i+width]))
    dist = np.array(dist)
    return dist
```

```python
slide_cov = [0]
for i in range(1, 300):
    slide_cov.append(np.cov(slide(mean_xy,lambda x:x.var(),width=i), out)[0,1])
plt.title('Covariance of slide width')
plt.plot(slide_cov)
```

```python
print('max cov:', np.argmax(slide_cov))
```

```python
slide_var_xy = slide(mean_xy,lambda x:x.var(),width=33)
plt.title('Var with slide width 33')
plt.plot(slide_var_xy)
```

```python
def binarization(data, threshold):
    dist = data.copy()
    dist[data < threshold]  = 0.
    dist[data >= threshold] = 1.
    return dist
```

```python
opt_threshold = []
for i in np.arange(0.0, 1.0, 0.001):
    opt_threshold.append( (binarization(slide_var_xy, i)==out).sum() )
plt.title('Threshold accuracy')
plt.plot(opt_threshold)
```

```python
print('max threshold:', np.argmax(opt_threshold)*0.001, '\nacc:', np.max(opt_threshold) / slide_var_xy.shape[0] )
```

```python
plt.plot( binarization(slide_var_xy, 0.024) )
```

```python
def get_leg_vibration(x, y):
    """
    貧乏ゆすりを検出する(精度: 91%程度)
    
    param: 
        x: ndarray型で加速度センサXの取得
        y: ndarray型で加速度センサYの取得
    return: 
        ndarray型で1なら貧乏ゆすり、0ならそれ以外
    """
    mean_xy = np.mean([x, y], axis=0)
    slide_var_xy = slide(mean_xy,lambda x:x.var(),width=33)
    return binarization(slide_var_xy, 0.024)
```

```python
plt.plot(out)
plt.plot(get_leg_vibration(x, y))
plt.title('acc(91%)')
```

```python
def noiseRemover(data, n):
    w = data.copy()
    for i in range(n):
        w = slide(w, lambda x: 0 if x[0]==1 and x[1]==0 or x[0]==0 and x[1]==1 else x[0] , width=2)
    for i in range(n):
        w = slide(w, lambda x: 1 if x[0]==1 and x[1]==0 or x[0]==0 and x[1]==1 else x[0] , width=2)
    return w
n = []
for i in range(50):
    n.append( (noiseRemover(get_leg_vibration(x, y), i)==out).sum() ) 
plt.plot(n)
```

```python
n_max = np.argmax(n)
print('max noize filter n 1番目:', n_max, ', acc:', np.max(n) / out.shape[0] )
```

```python
plt.plot( out )
plt.plot( noiseRemover(get_leg_vibration(x, y), n_max) )
plt.show()
```
